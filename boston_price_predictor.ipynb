```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
imputer.fit(housing)
```

```python
imputer.statistics_
#missing value for every colum
```

```python
x=imputer.transform(housing)
housingtr=pd.DataFrame(x,columns=housing.columns)
```

```python
housingtr.describe()
```

```python
# 3 objects estimators estimate parameter based on dataset ex imputer has fit and transform.fits the dataset and cal internal parameters
#transformers takes input returns output based on learning from fit() fit_transform() both
#predictors linear regression fit predict
```

```python
#pipeline
# feature scaling-minmax scaling(normalisation -value-min/max-min) and standardisation scaling-value-mean/std
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipe=Pipeline([('imputer',SimpleImputer(strategy="median")),('std_scaler',StandardScaler()),])
```

```python
housingnumtr=pipe.fit_transform(housingtr)
housingnumtr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model=DecisionTreeRegressor()
#model=LinearRegression()
model=RandomForestRegressor()
model.fit(housingnumtr,housing_labels)
```

```python
somd=housing.iloc[:5]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
housepred=model.predict(housingnumtr)
lin_mse=mean_squared_error(housing_labels,housepred)
lin_rmse=np.sqrt(lin_mse)
```

```python
lin_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housingnumtr,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmsescore=np.sqrt(-scores)
```

```python
rmsescore
```

```python
def print_scores(scores):
    print(scores,scores.mean(),scores.std())
print_scores(rmsescore)
```
